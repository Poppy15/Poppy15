```python

import matplotlib.pyplot as plt

from tabulate import tabulate

# Read data files
dfd = pd.read_csv("../data/raw/dfu_demographics_20240318.csv")
dfv = pd.read_csv("../data/raw/dfu_clinic_visits_20240318.csv")
dfb = pd.read_csv("../data/raw/dfu_bloods_20240318.csv")
dfc = pd.read_csv("../data/raw/dfu_conditions_20240318.csv")
imd = pd.read_csv("../data/external/Index_of_Multiple_Deprivation_(Dec_2019)_Lookup_in_England.csv")
dfmort = pd.read_csv("../data/raw/dfu_mortality_20240510.csv")
comorb = pd.read_pickle("../data/interim/vc_comorbdities_by_visit_needs_validation.pickle")

# Drop rows with missing values in 'person_id' or 'clinic_date' columns in dfv
dfv = dfv.dropna(subset=['person_id', 'clinic_date'])

# Drop rows with 'No matching concept' in 'ethnicity' column in dfd
dfd = dfd[dfd['ethnicity'] != 'No matching concept']

# Drop duplicates based on 'person_id' in dfd
dfd = dfd.drop_duplicates(subset=['person_id'])

# Drop rows with missing values in important columns in dfd
important_cols = ['person_id', 'year_of_birth', 'gender', 'ethnicity', 'lsoa']
dfd = dfd.dropna(subset=important_cols)

# Merge the cleaned demographic data with the visit data
df_merged = pd.merge(dfv, dfd[important_cols], on='person_id', how='inner')

# Keep only the specified columns
final_cols = ['person_id', 'clinic_date', 'year_of_birth', 'gender', 'ethnicity', 'lsoa']
dfd = df_merged[final_cols]

# Find the shape of the merged DataFrame
merged_shape = dfd.shape

# Display the shape of the merged DataFrame
print(f"The shape of the merged DataFrame is: {merged_shape}")

# Ensure clinic_date is in datetime format
dfd['clinic_date'] = pd.to_datetime(dfd['clinic_date'])

# Calculate the age at the time of each visit
dfd['age'] = dfd['clinic_date'].dt.year - dfd['year_of_birth']

# Define a dictionary to map ethnicities to their respective groups
ethnicity_groups = {
    'Asian Indian': 'Asian',
    'Indian': 'Asian',
    'Chinese': 'Asian',
    'Black': 'Black',
    'African': 'Black',
    'Pakistani': 'Pakistani',
    'White': 'White'
}

# Map the ethnicities to their groups
dfd['ethnicity_group'] = dfd['ethnicity'].map(ethnicity_groups)

# Drop rows with missing age values if any
dfd = dfd.dropna(subset=['age'])

# Aggregate the data to have one row per person_id
aggregated_dfd = dfd.groupby('person_id').agg({
    'clinic_date': ['min', 'max'],
    'year_of_birth': 'first',
    'gender': 'first',
    'ethnicity': 'first',
    'lsoa': 'first',
    'age': 'mean',
    'ethnicity_group': 'first'
}).reset_index()

# Rename the columns to simpler names
aggregated_dfd.columns = ['person_id', 'first_clinic_date', 'last_clinic_date', 'year_of_birth', 'gender', 'ethnicity', 'lsoa', 'mean_age', 'ethnicity_group']

# Display the aggregated DataFrame
print(aggregated_dfd.head())

# Calculate statistics for ethnicity, gender, and age
ethnicity_stats = aggregated_dfd.groupby('ethnicity_group')['mean_age'].describe().reset_index()
gender_stats = aggregated_dfd.groupby('gender')['mean_age'].describe().reset_index()
ethnicity_gender_stats = aggregated_dfd.groupby(['ethnicity_group', 'gender'])['mean_age'].describe().reset_index()

# Display the statistics in formatted tables
print("Ethnicity Statistics:")
print(tabulate(ethnicity_stats, headers='keys', tablefmt='pretty', showindex=False))

print("\nGender Statistics:")
print(tabulate(gender_stats, headers='keys', tablefmt='pretty', showindex=False))

print("\nEthnicity and Gender Statistics:")
print(tabulate(ethnicity_gender_stats, headers='keys', tablefmt='pretty', showindex=False))

# Group by LSOA and count the number of visits
visits_per_area = dfd['lsoa'].value_counts()

# Find the LSOA with the maximum number of visits
most_visited_area = visits_per_area.idxmax()
max_visits = visits_per_area.max()

# Print the result
print(f"The most visited area is {most_visited_area} with {max_visits} visits.")

# Create a pie chart to visualize the distribution of visits across different areas
plt.figure(figsize=(10, 8))
visits_per_area.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Visits Across Areas')
plt.ylabel('')  # Remove the ylabel
plt.show()
```

```python
#To find the first 7 top visited areas (LSOA) and the number of patients in each area, you can follow these steps:

#1.Group by LSOA and count the number of visits.
#2.Identify the top 7 LSOAs based on visit count.
#3.Count the number of unique patients for each of the top 7 LSOAs.


# Find the top 7 most visited areas and the number of patients in each area
top_7_areas = visits_per_area.head(7).index
top_7_visits = dfd[dfd['lsoa'].isin(top_7_areas)]

# Count the number of unique patients for each of the top 7 areas
patients_per_top_area = top_7_visits.groupby('lsoa')['person_id'].nunique().reset_index()
patients_per_top_area.columns = ['lsoa', 'unique_patients']

# Display the top 7 areas with visit counts and unique patient counts
top_7_stats = visits_per_area.head(7).reset_index()
top_7_stats.columns = ['lsoa', 'visit_count']
top_7_stats = pd.merge(top_7_stats, patients_per_top_area, on='lsoa')

print("\nTop 7 Most Visited Areas and Unique Patients Count:")
print(tabulate(top_7_stats, headers='keys', tablefmt='pretty', showindex=False))
```

```python
# Function to get patient details
def get_patient_details(person_id):
    # Filter for the specific patient in the visits data
    patient_visits = dfv[dfv['person_id'] == person_id]

    # Filter for the specific patient in the demographics data
    patient_demo = aggregated_dfd[aggregated_dfd['person_id'] == person_id]

    if patient_visits.empty or patient_demo.empty:
        return f"No data available for patient ID: {person_id}"
    
    # Merge patient visits with demographic data
    patient_details = pd.merge(patient_visits, patient_demo, on='person_id', how='inner')

    # Display the patient's visit and demographic details
    print(f"Details for patient ID: {person_id}")
    print(tabulate(patient_details, headers='keys', tablefmt='pretty', showindex=False))

# Example usage: Replace '123456' with the actual person_id you want to look up
person_id = 9811661
get_patient_details(person_id)
```

# Blood Test merge to Demographics and Visits 

Merge blood measurements (dfb) with demographics on 'person_id' and 'measurement_date' to include blood measurements in the demographic information.
we need 'creatinine_tests' and 'Haemoglobin A1c level' columns to merge with 'dfd'

```python
import pandas as pd

# Assuming dfb and dfd are already loaded

# Drop duplicates based on 'person_id', 'measurement_date', and 'concept_name'
dfb_cleaned = dfb.drop_duplicates(subset=['person_id', 'measurement_date', 'concept_name'])

# Drop rows with missing or zero values in 'value_as_number' column
dfb_cleaned = dfb_cleaned[dfb_cleaned['value_as_number'].notna() & (dfb_cleaned['value_as_number'] != 0)]

# Select the required columns
dfb_cleaned = dfb_cleaned[['person_id', 'measurement_date', 'measurement_concept_id', 'concept_name', 'value_as_number']]

# Display the first few rows of the cleaned DataFrame
print("Cleaned DataFrame:")
print(dfb_cleaned.head())

# Define relevant tests
relevant_tests = [
    'Serum creatinine level', 
    'Estimated glomerular filtration rate by laboratory calculation', 
    'Haemoglobin A1c level'
]

# Filter dfb_cleaned to include only relevant tests
filtered_dfb = dfb_cleaned[dfb_cleaned['concept_name'].isin(relevant_tests)]

# Display the first few rows of the filtered DataFrame
print("Filtered DataFrame:")
print(filtered_dfb.head())

# Aggregate blood test results by patient to ensure unique person_id
aggregated_dfb = filtered_dfb.groupby(['person_id', 'concept_name'])['value_as_number'].mean().unstack().reset_index()

# Debug: Print the columns of aggregated_dfb to confirm column names
print("Aggregated DataFrame Columns:")
print(aggregated_dfb.columns)

# Merge aggregated blood test results with demographic data on 'person_id'
merged_dfb = pd.merge(dfd, aggregated_dfb, on='person_id', how='inner')

# Ensure person_id is unique in the final merged DataFrame
merged_dfb = merged_dfb.drop_duplicates(subset=['person_id'])

# Correct the column name if there's a mismatch
# Print the columns of merged_dfb to confirm the names
print("Merged DataFrame Columns Before Selection:")
print(merged_dfb.columns)

# Select only the columns of interest
columns_of_interest = [
    'person_id', 'clinic_date', 'year_of_birth', 'gender', 'ethnicity', 
    'lsoa', 'age', 'ethnicity_group', 'Haemoglobin A1c level', 
    'Serum creatinine level', 'Estimated glomerular filtration rate by laboratory calculation'
]

# Ensure all columns of interest are in the DataFrame
existing_columns_of_interest = [col for col in columns_of_interest if col in merged_dfb.columns]

# Select the available columns
merged_dfb = merged_dfb[existing_columns_of_interest]

# Display the merged DataFrame
print("Merged DataFrame:")
print(merged_dfb.head())

# Optional: Display count of all available blood tests and how many patients they are available for
blood_test_counts = pd.concat([dfb_cleaned['concept_name'].value_counts(), dfb_cleaned.groupby('concept_name')['person_id'].nunique()], axis=1)
blood_test_counts.columns = ['Test Count', 'Patients Count']
print("Blood Test Counts:")
print(blood_test_counts)

# Additional filtered dataframes from Code 1 (optional)
creatinine_tests = dfb_cleaned[dfb_cleaned['concept_name'].str.contains('Creatinine', case=False)]
Haemoglobin_A1c_level_tests = dfb_cleaned[dfb_cleaned['concept_name'].str.contains('Haemoglobin A1c level', case=False)]
egfr_tests = dfb_cleaned[dfb_cleaned['concept_name'].str.contains('Estimated glomerular filtration rate by laboratory calculation', case=False)]

# Display the first few rows of the additional filtered DataFrames (optional)
print("Creatinine Tests:")
print(creatinine_tests.head())
print("Haemoglobin A1c Level Tests:")
print(Haemoglobin_A1c_level_tests.head())
print("eGFR Tests:")
print(egfr_tests.head())
```

```python
#Average Serum Creatinine Levels Across Different Age Groups and Genders

# Define age groups
bins = [0, 18, 35, 50, 65, 80, np.inf]
labels = ['0-18', '19-35', '36-50', '51-65', '66-80', '80+']
merged_dfb['age_group'] = pd.cut(merged_dfb['age'], bins=bins, labels=labels, right=False)

# a) Average Serum Creatinine Levels Across Different Age Groups and Genders
avg_creatinine_by_age_gender = merged_dfb.groupby(['age_group', 'gender'])['Serum creatinine level'].mean().reset_index()

# Display the results
print("Average Serum Creatinine Levels by Age Group and Gender:")
print(avg_creatinine_by_age_gender)

# Visualize the results
sns.barplot(data=avg_creatinine_by_age_gender, x='age_group', y='Serum creatinine level', hue='gender')
plt.xlabel('Age Group')
plt.ylabel('Average Serum Creatinine Level')
plt.title('Average Serum Creatinine Levels by Age Group and Gender')
plt.show()

```

```python
# b) Correlation of Serum Creatinine Levels with Demographic Factors (Ethnicity and Age)
# Correlation matrix for numerical demographics
demographics_df = merged_dfb[['age', 'Serum creatinine level']]
correlation_matrix = demographics_df.corr()

# Display the correlation matrix
print("Correlation Matrix for Age and Serum Creatinine Level:")
print(correlation_matrix)

# Visualize the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix for Age and Serum Creatinine Level')
plt.show()

# Boxplot to visualize the relationship between serum creatinine levels and ethnicity
plt.figure(figsize=(12, 6))
sns.boxplot(data=merged_dfb, x='ethnicity', y='Serum creatinine level')
plt.xticks(rotation=90)
plt.xlabel('Ethnicity')
plt.ylabel('Serum Creatinine Level')
plt.title('Serum Creatinine Levels by Ethnicity')
plt.show()
```
